```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = "225b370723956e3ae4784f52e2800bff"

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
# List of city data
city_list = []
city_info = []

# Print on log 
print("Begin")
print(".........................")


# Loop through the list of cities and perform a request for data on each
for city in cities:
    print(f"Processing record for the city: {city}")   
    response = requests.get(query_url + city).json()
    searchCityID = response.get("id")
    if response.get("id"):
        print(f"     Record found for city: {city}    city id: {searchCityID}")
        city_info.append(response)
        city_list.append(city)
        
    else:
        print(f"     No weather record found for city: {city}")
          
# Indicate that Data Loading is complete 
print("-----------------------------")
print("Complete")
```

```python
len(city_info)
```

```python

# Use list comprehension to gather necessary data series

lat_data = [data.get("coord").get("lat") for data in city_info]
max_temp = [data.get("main").get("temp") for data in city_info]
humidity = [data.get("main").get("humidity") for data in city_info]
clouds = [data.get("clouds").get("all") for data in city_info]
wind = [data.get("wind").get("speed") for data in city_info]
```

```python
# Convert jsons into dataframe
weather_data = pd.DataFrame({"City Name":city_list,
                            "Latitude":lat_data,
                            "Temperature":max_temp,
                            "Humidity":humidity,
                            "Cloudiness":clouds,
                            "Windspeed":wind})


# Force column order
weather_data = weather_data[["City Name", "Latitude","Temperature","Humidity","Cloudiness","Windspeed"]]
weather_data.head()
```

```python
# Save weather data to csv file
weather_data.to_csv("weatherdata.csv", encoding="utf-8", index=False)
```
